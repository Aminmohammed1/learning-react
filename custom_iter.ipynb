```python
class countdown:

    def __init__(self, num):
        self.num = num

    def __iter__(self):
        return self

    def __next__(self):
        if(self.num < 0):
            raise StopIteration("The counter has reached 0")
        self.num -= 1 # 4 3 2 1 0
        return self.num + 1 # 5 4 3 2 1 0
    
# c = countdown(5)
for i in countdown(5):
    print(i)

```

```python
class MyRange:
    
    def __init__(self, end, start = 0, step= 1):
        end, start = start, end
        self.start = start
        self.end = end
        self.step = step

    def __iter__(self):
        return self

    def __next__(self):
        if (self.start >= self.end):
            raise StopIteration
        self.start += self.step
        return self.start - self.step


        # for i in range(0, 2, 5)

for i in MyRange(0, 10, 2):
    print(i)

for i in range(0, 10, 2):
    print(i)    
```

```python
 # 0 1 1 2 3 5 8 13 21
def fibonacci(num):
    print("got called...")
    a,b = 0, 1
    # if num < =0 :
        # return num
    for i in range(num):
        yield a
        a, b = b, a+b

for i in fibonacci(-5):
    print(i)
# print(fibonacci(5))

```

```python
def evenGen(num):
    for i in range(1, num+1):
        if (i % 2 == 0):
            yield i

def squareGen(num):
    for i in num:
        yield i*i

# n = 

# n = squareGen(evenGen(10))

for i in squareGen(evenGen(10)):
    print(i)
```
